```python
x=np.arange(-5,5,1)
y0=x**2
y1=x**3
plt.plot(x,y0,label="Sq")
plt.plot(x,y1,label="Cube")
plt.legend(loc="upper center",frameon=False)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
